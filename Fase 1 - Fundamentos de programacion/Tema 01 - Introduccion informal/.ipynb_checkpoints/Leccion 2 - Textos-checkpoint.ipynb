```python
# Podemos utilizar comillas simples
'Hola mundo'
```

```python
# O comillas dobles
"Hola mundo"
```

```python
# Al utilizar ambas podemos incluirlas en el propio texto
'Este texto incluye unas " " '
```

```python
# Vemos cómo incluir las comillas en un texto
"Esta 'palabra' se encuentra escrita entre comillas simples"
```

```python
# También podemos utilizar \ como escape para que no se tenga en cuenta sino como texto
"Esta \"palabra\" se encuentra escrita entre comillas dobles"
```

```python
# Utilizando el escape del intérprete podemos utilizar \t para el tabulador
print('Un texto\tUna tabulación')
```

```python
# Utilizando el escape del intérprete podemos utilizar \n para nueva línea
print('Un texto\nUn salto de línea')
```

```python
# Podemos indicar que el texto se interprete TAL CUAL, es decir, en modo RAW, para que no interprete ningún caracter especial
print(r'C:\nombre\directorio')
```

```python
# Podemos utilizar las triples comillas para que se interpreten automaticamente saltos de linea
print('''Una línea
Otra línea
Otra línea''')
```

```python
# Podemos guardar texto en variables
c = "Esto es una cadena\ncon dos líneas"
```

```python
# OJO hay que utilizar print() para que se interpreten los caracteres especiales
c
```

```python
print(c)
```

```python
# Concatenamos texto con el operador suma
c+c
```

```python
s = "Una cadena" "compuesta de dos cadenas"
print(s)
```

```python
s1 = "Una cadena"
s2 = "Otra cadena"
print(s1+s2)
```

```python
# Podemos multiplicar el texto
diez_espacios = " " * 10
print(diez_espacios + "Un texto a diez espacios")
```

```python
# Podemos utilizar los indices para acceder a caracteres
palabra = 'Python'
palabra[0] # Muestra el primer caracter del texto
```

```python
# El tercer caracter
palabra[2]
```

```python
# El último
palabra[-1]
```

```python
# Podemos utilizar el slicing para coger varios caracteres juntos
palabra[0:3]
```

```python
# Vemos que se incluye el primer caracter pero se excluye el último por tanto si queremos ver desde el tercer caracter hasta el último
palabra[2:]
```

```python
# El tipo cadena no es reasignable, es decir no podemos cambiar un caracter concreto del texto haciendo lo siguiente
palabra[0] = 'N'
```

```python
# Tendríamos que crear de nuevo la cadena de la siguiente forma
palabra = 'N' + palabra[1:]
print(palabra)
```

```python
# Hay una función muy útil que podemos utilizar también llamada len()
len(palabra)
```

```python
# Podemos voltear una cadena de caracteres utilizando el slicing añadiendo un tercer parámetro
palabra_volteada=palabra[::-1]
print(palabra_volteada)
```
