```python
# Si introducimos un texto se mostrará por pantalla a la hora de escribir por teclado el valor
valor = input('Introduce un valor: ')
```

```python
# Para poder utilizarlo como número
valor = input("Introduce un número entero: ")
```

```python
# Hacemos la siguiente transformación con la función int()
int(valor)
```

```python
# Ya podríamos utilizarlo como número
int(valor) + 100
```

```python
# Pero esto no sirve con decimales
valor = input("Introduce un número decimal: ")
```

```python
int(valor)
```

```python
# Al ser un decimal utilizamos la función float()
float(valor)
```

```python
# Podríamos hacer todo junto, por ejemplo
valor = float(input("Introduce un número entero o decimal"))
valor
```
