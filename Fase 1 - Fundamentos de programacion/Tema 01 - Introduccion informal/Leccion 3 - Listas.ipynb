```python
# En una lista SI que podemos asignar valores, por ejemplo
pares = [0,2,4,5,8,10]
```

```python
# Corregimos reasignando
pares[3]=6
pares
```

```python
# Además podemos añadir elementos a través de métodos
pares.append(12)
pares
```

```python
# También podemos hacer asignación por slicing
letras = ['a','b','c','d','e','f']
```

```python
letras[:3]=['A','B','C']
letras
```

```python
# También podemos borrar elementos
letras[:3] = []
letras
```
